## Geocoding the dataframe

Here we iterate over the rows of the dataframe and geocode each city, obtain it's latitude and longitude and create new columns to store them
      

```
for index, row in df2.iterrows():
    c=row["current_city"]
    geolocator = Nominatim(user_agent="specify_your_app_name_here")
    location = geolocator.geocode(c)
    #print(location.latitude)
    #x.loc[index,'State']=results[0]['admin1']
    if location:
      df2.loc[index,'latitude']=location.latitude
      df2.loc[index, 'longitude'] = location.longitude
    else:
      print(c)
      df2.loc[index,'latitude']=0
      df2.loc[index,'longitude']=0
```

**The code is written in such a way that, if the geocoder fails to identify a location of the city as a valid location, it get's printed out**

**As seen from the above cell of code: **


1.   **OPPS KAMALA MILLS MUMBAI**
2.   **GURGOAN**
3.   **MANALURPET**

were failed to be geocoded, therefore, they have been manually changed to their nearest recognizable city, or in some cases, their names have been changed according to the names on Google Maps.

***Ex - MANALURPET has been changed to  MANALURPETTAI***



```
df2=df2.replace({'OPPS KAMALA MILLS MUMBAI': 'MUMBAI', 'GURGOAN': 'GURGAON','MANALURPET':'MANALURPETTAI'})
```

**Below code segment to create new columns 'current_latitude' and 'current_longitude' for the current_city**

```
for index, row in df2.iterrows():
    c=row["current_city"]
    #d=row["permanent_city"]
    geolocator = Nominatim(user_agent="specify_your_app_name_here")
    location = geolocator.geocode(c)
   # location1= geolocator.geocode(d)
    #print(location.latitude)
    #x.loc[index,'State']=results[0]['admin1']
    if location:
      df2.loc[index,'current_latitude']=location.latitude
      df2.loc[index, 'current_longitude'] = location.longitude
    else:
      print(c)
      df2.loc[index,'current_latitude']=0
      df2.loc[index,'current_longitude']=0
```

**Now doing the same operation for Permanent City **
**With this we have new features, 'permanent_latitude' and 'permanent_longitude'**

```
for index, row in df2.iterrows():
    c=row["permanent_city"]
    d=row["permanent_state"]
    geolocator = Nominatim(user_agent="specify_your_app_name_here")
    location = geolocator.geocode(c)
   # location1= geolocator.geocode(d)
    #print(location.latitude)
    #x.loc[index,'State']=results[0]['admin1']
    if location:
      df2.loc[index,'permanent_latitude']=location.latitude
      df2.loc[index, 'permanent_longitude'] = location.longitude
    else:
      print(c , d)

      df2.loc[index,'current_latitude']=0
      df2.loc[index,'current_longitude']=0
```
